```python
import cv2
import numpy as np

img = cv2.imread("people.jpg")
print(img)
print("hsjıjdsfkdfpgş")
```

```python
img.shape
```

```python
img_width = img.shape[1]
img_height = img.shape[0]
```

```python
img_width
```

```python
img_blob = cv2.dnn.blobFromImage(img, 1/255, (416,416), swapRB = True, crop = False )  #görüntüyü 4boyutlu tensöre çevirdik, modelimiz 416 ololanı indirdik
#crop false çünkü resmi kırpmasını istemiyoruz
```

```python
img_blob
```

```python
labels = ["person", "train", "laptop", "bicycle", "motorcycle", "airplane", "bus", "trafficlight", 
         "firehydrant", "stopsign", "dog", "horse", "sheep", "cow", "elephant", "bear", "umbrella", 
          "handbag", "bottle", "keyboard", "book", "clock", "vase", "book", "apple", "banana", "mouse",
          "train", "truck", "boat", "parkingmeter", "bench", "bird", "cat", "zebra", "giraffe", "backpack",
          "tie", "suitcase", "frisbee", "skis", "snowboard", "sportsball", "kite", "baseballbat", "basebaliglove",
          "skateboard", "surfboard", "tennisracket", "wineglass", "cup", "fork", "knife", "spoon", "bowl",
          "sandwich", "orange", "broccoli", "carrot", "hotdog", "pizza","donut","cake", "chair", "sofa",
          "pattedplant","bed", "diningtable", "toilet", "tvmonitor", "remote","cellphone", "mocrowave", "oven",
          "toaster", "sink", "refrigerator", "scissors", "teddybear", "hairdrier", "toothbrush"]
```

```python
colors = ["0,255,255", "0,0,255", "255,255,0", "0,255,0"]
colors = [np.array(color.split(",")).astype("int") for color in colors]
colors = np.array(colors)
colors = np.tile(colors,(18,1))  
#tile fonksiyonu colors matrisini aşağıya doğru 18 kere çoğğalttı yana doğru 1 yerine 2 olsaydı 2 kere çoğaltırdı
```

```python
colors
```

```python
model = cv2.dnn.readNetFromDarknet("model/yolov3.cfg", "model/yolov3.weights")
```

```python
layers = model.getLayerNames()
layers
```

```python
output_layers = [layers[layer[0]-1] for layer in model.getUnconnectedOutLayers()]
output_layers
```

```python
model.setInput(img_blob)
```

```python
detection_layers = model.forward(output_layers)
detection_layers
```

```python
for detection_layer in detection_layers:
    for object_detection in detection_layer:
        scores = object_detection[5:]
        predicted_id = np.argmax(scores)
        confidence = scores[predicted_id]
        
        if confidence > 0.40:
            label = labels[predicted_id]
            bounding_box = object_detection[0:4] * np.array([img_width, img_height, img_width, img_height ])
            (box_center_x, box_center_y, box_width, box_height) = bounding_box.astype("int")
            
            start_x = int(box_center_x - (box_width/2))
            start_y = int(box_center_y - (box_height/2))
            
            end_x = start_x + box_width
            end_y = start_y + box_height
            
            box_color = colors[predicted_id]
            box_color = [int(each) for each in box_color]
            
            label = "{}: {:.2f}%".format(label, confidence*100)
            print("predicted object {}".format(label)) #güven skorunu ekrana yazdırmak için
            
            cv2.rectangle(img, (start_x, start_y), (end_x, end_y), box_color, 1)
            cv2.putText(img, label, (start_x, start_y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, box_color, 1)
            
            
            
        
```

```python
cv2.imshow("Detection Window" , img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
